```python
from atlassian import Jira
```


 How to get server info with health check
<br>
jira = Jira(url="https://jira.example.com/", username="gonchik.tsymzhitov", password="********")<br>
log = logging.getLogger("com.gonchik.python.scripts.example")<br>
logging.basicConfig(level=logging.ERROR)<br>
projects = jira.get_all_projects()<br>
for project in projects:<br>
    project_key = project.get("key")<br>
    try:<br>
        value = jira.get_project_issue_security_scheme(project_key).get("name") or "None"<br>
    except Exception as e:<br>
        log.error(e)<br>
        value = "None"<br>
    print(project_key + " has issue security scheme " + value)
