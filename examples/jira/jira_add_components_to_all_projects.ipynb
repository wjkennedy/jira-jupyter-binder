```python
# coding=utf-8
from atlassian import Jira
```

```python
jira = Jira(url="http://localhost:8080/", username="jira-administrator", password="admin")
```

```python
components = ["Data Base", "HTML", "JavaScript"]
```


That example show how to create components on all existing projects, only skipping the one in a provided list
<br>
project_to_skip = ["SI", "SA", "ETA"]<br>
for i in jira.get_all_projects(included_archived=None):<br>
    if i["key"] in project_to_skip:<br>
        print("Skipping project {} ".format(i["key"]))<br>
    else:<br>
        for j in components:<br>
            print("Creating in project {} ".format(i["key"]))<br>
            comp = {"project": i["key"], "name": j}<br>
            jira.create_component(comp)<br>
            print("{} - component created ".format(comp.get("name")))
