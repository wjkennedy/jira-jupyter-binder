```python
# coding=utf-8
from atlassian import Jira
```

```python
jira = Jira(url="http://localhost:8080/", username="jira-administrator", password="admin")
```


That example show how to copy components from one project into another
<br>
DST_PROJECT = "PROJECT_B"<br>
SRC_PROJECT = "PROJECT_A"<br>
components = jira.get_project_components(SRC_PROJECT)<br>
for component in components:<br>
    data = {<br>
        "project": DST_PROJECT,<br>
        "description": component.get("description"),<br>
        "leadUserName": component.get("leadUserName"),<br>
        "name": component.get("name"),<br>
        "assigneeType": component.get("assigneeType"),<br>
    }<br>
    jira.create_component(data)<br>
    print("{} - component created ".format(component.get("name")))
