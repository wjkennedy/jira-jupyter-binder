def convert_group_into_users_in_role(project_key, role_id, group_name):<br>
    users = jira.get_all_users_from_group(group=group_name, limit=1000).get("values")<br>
    for user in users:<br>
        jira.add_user_into_project_role(project_key=project_key, role_id=role_id, user_name=user.get("name"))<br>
        print("{} added into role_id  {} in {}".format(user.get("name"), role_id, project_key))<br>
group_name_to_find = "old-developers"<br>
roles = jira.get_all_global_project_roles()<br>
projects = jira.get_all_projects(included_archived=True)<br>
for project in projects:<br>
    for role in roles:<br>
        members_of_role = jira.get_project_actors_for_role_project(project.get("key"), role.get("id"))<br>
        if not members_of_role:<br>
            continue<br>
        for member in members_of_role:<br>
            if member.get("type") == "atlassian-group-role-actor":<br>
                if member.get("name") == group_name_to_find:<br>
                    print("{} has {}".format(project.get("key"), role.get("name")))<br>
                    convert_group_into_users_in_role(project.get("key"), role.get("id"), group_name_to_find)
