```python
# coding=utf-8
from atlassian import Confluence
```


This example shows how to export pages 
<br>
confluence = Confluence(url="http://localhost:8090", username="admin", password="admin")<br>
def save_file(content, title):<br>
    file_pdf = open(title + ".pdf", "w")<br>
    file_pdf.write(content)<br>
    file_pdf.close()<br>
    print("Completed")<br>
if __name__ == "__main__":<br>
    label = "super-important"<br>
    pages = confluence.get_all_pages_by_label(label=label, start=0, limit=10)<br>
    for page in pages:<br>
        response = confluence.get_page_as_pdf(page["id"])<br>
        save_file(content=response, title=page["title"])
