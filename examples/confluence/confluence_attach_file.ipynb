```python
# coding=utf-8
"""
This is example to attach file with mimetype
```


<br>
import logging<br>
# https://pypi.org/project/python-magic/<br>
import magic<br>
from atlassian import Confluence<br>
logging.basicConfig(level=logging.DEBUG)<br>
confluence = Confluence(<br>
    url="http://localhost:8090",<br>
    username="admin",<br>
    password="admin",<br>
)<br>
def attach_file(page_title, file_location, file_name, mime_type, space):<br>
    page_id = confluence.get_page_by_title(space=space, title=page_title).get("id") or None<br>
    if page_id is None:<br>
        return 1<br>
    try:<br>
        confluence.attach_file(<br>
            filename=file_location, name=file_name, content_type=mime_type, page_id=page_id, space=space<br>
        )<br>
    except Exception:<br>
        return 1<br>
    return 0<br>
mime_type = magic.Magic(mime=True)<br>
file_location_with_page = "~/test/test_file.pdf"<br>
file_name = "So excited overview of report.pdf"<br>
title = "The page with attachments"<br>
space = "TST"<br>
content_type = magic.from_file(file_name, mime=True)<br>
attach_file(<br>
    file_location=file_location_with_page, file_name=file_name, mime_type=content_type, page_title=title, space=space<br>
)
