```python
# coding=utf-8
from atlassian import Confluence
```


This example shows how to export pages
<br>
confluence = Confluence(url="http://localhost:8090", username="admin", password="admin")<br>
# Set page property<br>
data = {"key": "newprp", "value": {"anything": "goes"}}<br>
print("SET")<br>
print(confluence.set_page_property(242793586, data))<br>
# # Get page property<br>
print("GET")<br>
print(confluence.get_page_property(242793586, "newprp"))<br>
# Update page property<br>
data = {<br>
    "key": "newprp",<br>
    "value": {"anything": "goes around"},<br>
    "version": {"number": 2, "minorEdit": False, "hidden": False},<br>
}<br>
print("UPDATE")<br>
print(confluence.update_page_property(242793586, data))<br>
# Delete page property<br>
print("DELETE")<br>
print(confluence.delete_page_property(242793586, "newprp"))
