```python
# coding=utf-8
from atlassian import Confluence
from atlassian.confluence import ApiError
```


 This example shows a way to get the real reason for an exception
<br>
try:<br>
    confluence = Confluence(<br>
        url="http://some_site_without_permission.com",<br>
        username="admin",<br>
        password="admin",<br>
    )<br>
    result = confluence.get_user_details_by_username(username="gonchik.tsymzhitov", expand="status")<br>
except ApiError as e:<br>
    print("FAILURE: {}, caused by {}".format(e, e.reason if e.reason is not None else "unknown reason"))<br>
else:<br>
    print(result)
