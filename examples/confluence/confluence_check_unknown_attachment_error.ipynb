```python
# coding=utf-8
from atlassian import Confluence
```


This example how to detect unknown-attachments errors
<br>
confluence = Confluence(url="http://localhost:8090", username="admin", password="admin", timeout=185)<br>
def get_all_pages_ids(space_key):<br>
    page_ids = []<br>
    limit = 50<br>
    flag = True<br>
    step = 0<br>
    while flag:<br>
        values = confluence.get_all_pages_from_space(space=space_key, start=step * limit, limit=limit)<br>
        step += 1<br>
        if len(values) == 0:<br>
            flag = False<br>
            print("Extracted all pages excluding restricts")<br>
        else:<br>
            for value in values:<br>
                page_ids.append(value.get("id"))<br>
    return page_ids<br>
def check_unknown_attachment_in_space(space_key):<br>
  


```python
    Detect errors in space
    :param space_key:
    :return:
    """
    page_ids = get_all_pages_ids(space_key)
    print("Start review pages {} in {}".format(len(page_ids), space_key))
    for page_id in page_ids:
        link = confluence.has_unknown_attachment_error(page_id)
        if len(link) > 0:
            print(link)
```

```python
if __name__ == "__main__":
    space_list = confluence.get_all_spaces()
    for space in space_list:
        print("Start review {} space".format(space["key"]))
        check_unknown_attachment_in_space(space["key"])
```
