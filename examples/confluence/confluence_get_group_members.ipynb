```python
# coding=utf-8
from atlassian import Confluence
from pprint import pprint
```


This example shows how to get all users from group e.g. group_name 
<br>
confluence = Confluence(url="http://localhost:8090", username="admin", password="admin")<br>
def get_all_members(group_name):<br>
    flag = True<br>
    i = 0<br>
    limit = 50<br>
    result = []<br>
    while flag:<br>
        response = confluence.get_group_members(group_name=group_name, start=i * limit, limit=limit)<br>
        if response and len(response):<br>
            i += 1<br>
            result.append(response)<br>
        else:<br>
            flag = False<br>
    return result<br>
if __name__ == "__main__":<br>
    group_name = "confluence-users"<br>
    pprint(get_all_members(group_name=group_name))
