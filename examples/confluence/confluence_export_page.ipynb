```python
# coding=utf-8
from atlassian import Confluence
```


This example shows how to export pages
<br>
confluence = Confluence(<br>
    url="https://test.atlassian.net/wiki",<br>
    username="admin",<br>
    password="api-key",<br>
    api_version="cloud",<br>
)<br>
if __name__ == "__main__":<br>
    space = "TEST"<br>
    page_title = "Test"<br>
    page_id = confluence.get_page_id(space, page_title)<br>
    content = confluence.export_page(page_id)<br>
    with open(page_title + ".pdf", "wb") as pdf_file:<br>
        pdf_file.write(content)<br>
        pdf_file.close()<br>
        print("Completed")
