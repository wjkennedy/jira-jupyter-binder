```python
# coding=utf-8
from atlassian import Bitbucket
```


This example shows how to create, get and delete tags
<br>
bitbucket = Bitbucket(url="http://localhost:7990", username="admin", password="admin")<br>
if __name__ == "__main__":<br>
    response = bitbucket.set_tag(<br>
        project_key="gonchik.tsymzhitov",<br>
        repository_slug="gonchik",<br>
        tag_name="test1",<br>
        commit_revision="ebcf5fdffa0",<br>
        description="Stable release",<br>
    )<br>
    print("Response after set_tag method")<br>
    print(response)<br>
    response = bitbucket.get_project_tags(project_key="INT", repository_slug="jira-plugins", tag_name="test1")<br>
    print("Retrieve tag")<br>
    print(response)<br>
    print("Remove tag")<br>
    bitbucket.delete_tag(project_key="INT", repository_slug="jira-plugins", tag_name="test1")
